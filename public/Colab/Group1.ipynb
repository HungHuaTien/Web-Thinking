```
每人小費金額 = 2.06

```


### 5. 實驗5：油漆計算機

向使用者詢問以英尺為單位的房間長度、寬度和高度，並詢問門和窗戶的數量。提供要油漆的總面積和應變所需的油漆總量（假設每扇門可以扣除20平方英尺、每扇窗扣除15平方英尺，且油漆面積是每加侖350平方英尺。)

**虛擬程式碼**

從使用者取得高度、寬度、長度

從使用者取得窗戶和門的數量

設牆面積為(2*長*高)+(2*寬*高)

設無油漆面積為20*門數+15*窗戶數

設油漆面積為牆面積-無油漆面積

顯示油漆面積

設加侖數為牆面積/350

顯示四捨五入到百分位的加侖數

**範例執行**

以英尺為單位輸入房間長度：24

以英尺為單位輸入房間寬度：14

以英尺為單位輸入房間高度：9

輸入門數量：2

輸入窗戶數量: 4

需油漆總面積: 584

需油漆加侖數：167


## 應用設計專題25分：打造你的聊天機器人

請你根據本序列單元的專題：打造你的聊天機器人，設計一個你們的聊天機器人。

也就是說，請同學先寫腳本，人與機器人的互動內容；然後把這段腳本轉換成Python程式，就變成你們打造的聊天機器人了！
